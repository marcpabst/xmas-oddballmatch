```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
A = 440
B = 449

tones1 = [A, A, A, B, A, A, A, A, A, A, B, A, A, A, A, B, A, A, A, B, A, A, A, A, A]
soa1 = range(0, len(tones1)*100, 100)

tones2 = [A, A, A, A, B, A, A, A, A, B, A, A, A, A, A, A, A, A, A, B, A, A, A, A, B]
soa2 = range(0, len(tones2)*100, 100)

xmargin = 30
ymargin = 3

from matplotlib.gridspec import GridSpec
import matplotlib
cmap = matplotlib.cm.get_cmap('Set2')

color0 = cmap(3)
color1 = cmap(4)
color2 = cmap(3)

def add_rect(ax, start, len, color = "grey", **kwargs):
    ax.add_patch(patches.Rectangle((start*100 - xmargin, A - ymargin), len*100 + xmargin*2, 9 + ymargin*2, facecolor="None", edgecolor=color, linewidth=2,  **kwargs))

fig = plt.figure(figsize=(20,4))

gs = GridSpec(2, 3, width_ratios=[.03, 1,.03], wspace=0)

# Top panel
ax1 = fig.add_subplot(gs[0,1])
ax1.set_facecolor( (color0[0],color0[1],color0[2],.25) )

ax1.set_yticks([440, 449])
ax1.set_yticklabels( ["A", "B"], fontsize=15)

ax1.set_xticks(soa1)
ax1.set_xticklabels([])
ax1.tick_params(axis="x",direction="in")


ax1.set_ylim(A-5, B+5)
ax1.set_xlim(-50, len(soa2)*100-50)

ax1.plot(soa2, tones2, linestyle='None', marker="s", markersize=10, color="black", markerfacecolor=color0, markeredgewidth=1.5)

add_rect(ax1, 0, 4)
add_rect(ax1, 5, 4)
add_rect(ax1, 10, 4, linestyle="dashed")
add_rect(ax1, 15, 4)
add_rect(ax1, 20, 4)


# Bottom panel
ax2 = fig.add_subplot(gs[1,1])
ax2.set_facecolor( (color1[0],color1[1],color1[2],.25) )

ax2.set_yticks([440, 449])
ax2.set_yticklabels( ["A", "B"], fontsize=15)

ax2.set_xticks(soa2)
ax2.set_xticklabels([])
ax2.tick_params(axis="x",direction="in")

ax2.set_ylim(A-5, B+5)
ax2.set_xlim(-50, len(soa2)*100-50)

ax2.plot(soa1, tones1, linestyle='None', marker="s", markersize=10, color="black", markerfacecolor=color1, markeredgewidth=1.5)

add_rect(ax2, 10, 5)
add_rect(ax2, 19, 5, linestyle="dashed")

#ax2.set_xlabel("" fontsize=15)

# Top pannel title
ax1a = fig.add_subplot(gs[0,0])
ax1a.set_facecolor("lightgray")

ax1a.get_xaxis().set_visible(False)
ax1a.get_yaxis().set_visible(False)

ax1a.text(.5, .27, "A", color="black", ha="center", va="center", fontsize=15, transform=ax1a.transAxes)
ax1a.text(.5, .73, "B", color="black", ha="center", va="center", fontsize=15, transform=ax1a.transAxes)


ax1t = fig.add_subplot(gs[0,2])
ax1t.get_xaxis().set_visible(False)
ax1t.get_yaxis().set_visible(False)
ax1t.set_facecolor( (color0[0],color0[1],color0[2],1) )

ax1t.text(.5, .5, "pred", color="white", ha="center", va="center", fontsize=15, rotation = 90, transform=ax1t.transAxes, fontweight='bold')

# Bottom pannel title
ax2a = fig.add_subplot(gs[1,0])
ax2a.set_facecolor("lightgray")

ax2a.get_xaxis().set_visible(False)
ax2a.get_yaxis().set_visible(False)

ax2a.text(.5, .27, "A", color="black", ha="center", va="center", fontsize=15, transform=ax2a.transAxes)
ax2a.text(.5, .73, "B", color="black", ha="center", va="center", fontsize=15, transform=ax2a.transAxes)


ax2t = fig.add_subplot(gs[1,2])
ax2t.get_xaxis().set_visible(False)
ax2t.get_yaxis().set_visible(False)
ax2t.set_facecolor( (color1[0],color1[1],color1[2],1) )

ax2t.text(.5, .5, "rand", color="white", ha="center", va="center", fontsize=15, rotation = 90, transform=ax2t.transAxes, fontweight='bold')


#ax1.set_ylim(A-5, B+5)
#ax1.set_xlim(-50, len(soa1)*100-50)

fig.savefig("/home/marc/ba-thesis/input/figures/fig1.pdf", bbox_inches='tight')
fig.savefig("/home/marc/ba-thesis/input/figures/fig1.png", bbox_inches='tight')
```
