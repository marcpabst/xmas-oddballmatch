```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as patches
```

```python
A = 440
B = 449

tones1 = [B, A, A, A, A, B, A, A, A, A, A, A, B, A, A, A, A, A, A, A, B, A, A, A, A, A]
soa1 = range(0, len(tones1)*100, 100)

A = 440 + 20
B = 449 + 20

tones2 = [B, A, A, A, A, B, A, A, A, A, B, A, A, A, A, B, A, A, A, A, B, A, A, A, A, A]
soa2 = range(0, len(tones2)*100, 100)

xmargin = 30
ymargin = 2

matplotlib.rcParams['font.size'] = 6


from matplotlib.gridspec import GridSpec
import matplotlib
cmap = matplotlib.cm.get_cmap('Set2')

color0 = cmap(3)
color1 = cmap(4)
color2 = cmap(3)

def add_rect(ax, start, len, color = "grey", **kwargs):
    ax.add_patch(patches.Rectangle((start*100 - xmargin, A - 20 - ymargin), len*100 + xmargin*2, 9 + 20 +ymargin*2, facecolor="None", edgecolor=color, linewidth=1,  **kwargs))

fig = plt.figure(figsize=(6.25,1.5))

gs = GridSpec(1, 3, width_ratios=[.03, 1,.03], wspace=0, left=0,top=1,right=1,bottom=0)

# Top panel
ax1 = fig.add_subplot(gs[0,1])
#ax1.set_facecolor( (color0[0],color0[1],color0[2],.25) )

ax1.set_yticks([440, 449, A, B])
ax1.set_yticklabels( ["A", "B", "A", "B"])

ax1.set_xticks(soa1)
ax1.set_xticklabels([])


ax1.spines['top'].set_visible(True)
ax1.spines['right'].set_visible(True)
ax1.spines['left'].set_visible(True)
ax1.spines['bottom'].set_visible(True)

# ax1.spines['left'].set_bounds(B,A)
# ax1.spines['bottom'].set_bounds(soa1[0], soa1[-1])

ax1.set_ylim(A-20-5, B+5)
ax1.set_xlim(-100, len(soa2)*100)


ax1.tick_params(axis="x",direction="in", pad=-22)

ax1.plot(soa2, tones2, linestyle='None', marker="s", markersize=4, color="black", markerfacecolor=color0, markeredgewidth=.5)

ax1.plot(soa1, tones1, linestyle='None', marker="s", markersize=4, color="black", markerfacecolor=color1, markeredgewidth=.5)


add_rect(ax1, 1, 4, linestyle="dotted")
add_rect(ax1, 21, 4, linestyle="dotted")




fig.savefig("/media/marc/Medien/xmas-oddballmatch/ba-thesis/input/figures/fig_tones.pdf", bbox_inches='tight')
fig.savefig("/media/marc/Medien/xmas-oddballmatch/ba-thesis/input/figures/fig_tones.png", dpi = 120,  bbox_inches='tight')
```
